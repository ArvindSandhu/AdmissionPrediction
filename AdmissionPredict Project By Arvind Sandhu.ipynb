```python
df.rename(columns = {'Chance of Admit ':'Chance of Admit', 'LOR ':'LOR'}, inplace=True)
```

```python
X = df.drop(["Chance of Admit"],axis=1)
y = df["Chance of Admit"].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 101)
```

```python
# normalization
from sklearn.preprocessing import MinMaxScaler
scalerX = MinMaxScaler(feature_range=(0, 1))
X_train[X_train.columns] = scalerX.fit_transform(X_train[X_train.columns])
X_test[X_test.columns] = scalerX.transform(X_test[X_test.columns])
```

```python
y_train_01 = [1 if each > 0.83 else 0 for each in y_train]
y_test_01  = [1 if each > 0.83 else 0 for each in y_test]

# list to array
y_train_01 = np.array(y_train_01)
y_test_01 = np.array(y_test_01)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logr = LogisticRegression()
logr.fit(X_train,y_train_01)
```

```python
y_predlogr = logr.predict(X_test)
```

```python
from sklearn import metrics
from sklearn.metrics import accuracy_score
print("Accuracy Score:", accuracy_score(y_predlogr, y_test_01))
```

```python
#Accuracy score on Test and Train
from sklearn.metrics import accuracy_score, recall_score, roc_auc_score

print("\nAccuracy score: %f" %(accuracy_score(y_test_01,y_predlogr) * 100))
print("Recall score : %f" %(recall_score(y_test_01,y_predlogr) * 100))
print("ROC score : %f\n" %(roc_auc_score(y_test_01,y_predlogr) * 100))
```

<h1>Observation </h1>
<h5>1. Here, recall depecits the the ratio of correctly
predicted positive observations to the all observations in actual class whose value is 0.9 in this case, which is pretty good.
<h5>2. ROC represents the area under the roc curve. It shows the performance of a classification model at all classification thresholds and demonstrates two parameters: True Positives and False Positives. Higher the area under curve, better the model.


```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test_01,y_predlogr))
cm_lrc = confusion_matrix(y_test_01,y_predlogr)
```
